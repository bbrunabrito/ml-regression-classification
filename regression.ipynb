# Regressão Linear 

```python
import numpy as np
import matplotlib.pyplot as plt
```


## Regressão Linear utilizando o Método dos Mínimos Quadrados

```python
X = 2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.rand(100,1)

plt.plot(X, y, "b.")
plt.axis([0,2,0,15])
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)), X]
```

```python
theta_param = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_param
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_param)

plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

## Regressão Linear com Gradiente Descendente

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
n = 0.1 #taxa de aprendizado
iterations = 1000
m = 100 #nº de instâncias 
```

```python
X_b = np.c_[np.ones((100, 1)), X]
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
theta = np.random.randn(2, 1) #inicialização aleatória
```

```python
for i in range(iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - n * gradients
```
